## Installation

* `curl -O https://dl.google.com/dl/cloudsdk/channels/rapid/downloads/google-cloud-cli-linux-arm.tar.gz`
* `tar -xvzf google-cloud-sdk-*.tar.gz`
* `cd google-cloud-sdk`
* `./install.sh`

## Initialization and Login

* `gcloud init`
* `gcloud auth application-default login`
